### Supplying a dataset's contextual description

All of the celestial bodies that revolve around the Sun are collectively referred to as the Solar System. Some 4.6 billion years ago, it was born when a massive interstellar molecular cloud disintegrated under the force of gravity. The Sun still accounts for almost all of the mass in the solar system (98.18 percent), but Jupiter possesses most of the rest. Mercury, Venus, Earth, and Mars are all considered terrestrial planets since they are mostly composed of rock and metal. In addition to Earth, there are three more massive planets in the solar system. The two largest planets, Jupiter and Saturn, are known as gas giants because they are made up mostly of hydrogen and helium; the next two largest planets, Uranus and Neptune, are known as ice giants because they are made up mostly of volatile substances with relatively high melting points compared to hydrogen and helium, such as water, ammonia, and methane. All eight planets move in nearly circular paths that cross the ecliptic plane or go in the opposite direction.

## Basic visualizations

### Bar graph 

```python
df.plot(kind="bar",
       x="Planet",
       y="Diameter (km)")
plt.title("Planet diameter")
plt.show()
```

That is a bar chart, by the way. The planets were shown along the x-axis, while the y-axis was the diameter in kilometers. For this purpose, the color blue works well since it allows us to clearly discern the planets and their diameters in kilometers.

Bar graph showing the relative diameters of the planets, with Jupiter on top, Saturn in second place, and Uranus in third place, in terms of kilometers. When compared to Neptune, Uranus is much larger in diameter. The diameter of Mercury is the smallest of the planets, followed by Mars and Venus.

In terms of planet size, Earth and Neptune are somewhat in the middle.

I believe a rectangular form would work best for the markers in the bar graph to indicate these planetary characteristics. Because of this, determining the planetary diameters is much less of a guessing game. 

### Scatter plot 

```python
df.plot(kind="scatter",
       x="Diameter (km)",
       y="Mean Temperature (C)")
plt.title("Mean temperature vs. diameter")
plt.show()
```

This graphic represents a scatter plot. I opted for this chart instead of a line graph to demonstrate the correlation between mean temperature in C and diameter in kilometers. The x-axis represents the mean temperature in degrees Celsius, while the y-axis represents the diameter of the planets in kilometers.

The scatter plot suggests that there is no relationship between the two factors. Nonetheless, the fact that smaller planets tend to be hotter than bigger ones is an intriguing observation. I am not an astronomer, but this helps explain why Earth is so large relative to other planets by considering their average temperatures. If this is right, then Jupiter must be the coldest planet in the solar system. 

As for the symbols used in the scatter plot, circles work well for representing the points that stand for the average temperatures in degrees Celsius and the planetary sizes in kilometers. The scatter graph is more legible.
